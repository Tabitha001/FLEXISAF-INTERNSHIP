```python
#A program to check if input number is prime number

number = int(input('Enter a number:'))

if number > 1:
    for i in range(2, number):
        if number % i == 0:
            print(number, 'is not a prime number')
            break
    else:
        print(number, ' is a prime number')
```

```python
#A program that prints the numbers from 1 to 100.
#But for multiples of three print “Fizz” instead of the number and for the multiples of five print “Buzz”.
#For numbers which are multiples of both three and five print “FizzBuzz”.


for i in range (1, 101):
    if i % 3 == 0 and i % 5 == 0:
      print('FizzBuzz')
    elif i % 3 == 0:
      print ('Fizz')
    elif i % 5 == 0:
      print ('Buzz')
    else:
      print(i)
```

```python
#program to create function func1() to accept a variable length of arguments and print their value.

def func1(*args):
    for i in args:
        print(i)


#call function
func1(7, 6, 1, "hey there", [8, 5, 4])
```

```python
#a program to create a recursive function to calculate the sum of numbers from 0 to 15.
def recursive_sum(n):

  if n == 0:
    return 0
  else:
    return n + recursive_sum(n-1)

# Calculate the sum from 0 to 15
result = recursive_sum(15)

# Print the result
print("The sum of numbers from 0 to 15 is:", result)
```

```python
#import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/company_sales_data.csv')
data.head()
```

```python
#Read Total profit of all months and show it using a line plot.
total_profit = data['total_profit']
total_profit
```

```python
plt.plot(data['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.title('Company Sales Data')
plt.show
```

```python
bathingsoap_and_facewash = data[['month_number', 'bathingsoap', 'facewash']]
bathingsoap_and_facewash
```

```python
#Read Bathing soap and facewash of all months and display it using the Subplot

# Create a figure with 1 row and 2 columns
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot data for Bathing Soap Sales
ax1.plot(data['month_number'], data['bathingsoap'], color="green", marker="o", linestyle="--", label="Bathing Soap Sales")
ax1.set_title("Bathing Soap Sales per Month", color="blue")
ax1.set_xlabel("Months")
ax1.set_ylabel("Sales")
ax1.grid(True)

# Plot data for Facewash Sales
ax2.plot(data['month_number'], data['facewash'], color="red", marker="s", linestyle=":", label="Facewash Sales")
ax2.set_title("Facewash Sales per Month", color="blue")
ax2.set_xlabel("Months")
ax2.set_ylabel("Sales")
ax2.grid(True)
```
